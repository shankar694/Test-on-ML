```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

```

```python
Train_data= pd.read_csv("/content/drive/MyDrive/Topmentor/train.csv")
Train_data.head()
```

```python
Test_data = pd.read_csv("/content/drive/MyDrive/Topmentor/test.csv")
Test_data.head()
```

```python
Train_data['datetime'] = pd.to_datetime(Train_data['datetime'], format='%d-%m-%Y %H:%M', errors='coerce')
Train_data['hour'] = Train_data['datetime'].dt.hour
Train_data['day'] = Train_data['datetime'].dt.day
Train_data['month'] = Train_data['datetime'].dt.month
Train_data['year'] = Train_data['datetime'].dt.year
Train_data['month'] = Train_data['datetime'].dt.month
Train_data['year'] = Train_data['datetime'].dt.year
```

```python
Train_data['weekday'] = Train_data['datetime'].dt.dayofweek
```

```python
X = Train_data[['season', 'holiday', 'workingday', 'weather', 'temp', 'atemp', 'humidity', 'windspeed', 'hour', 'day', 'month', 'year', 'weekday']]
y = Train_data['count']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=123)

model = RandomForestRegressor(n_estimators=100, random_state=123)
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_val)
```

```python
pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_squared_log_error
```

```python
r2_score(y_val, predictions)
```

```python
mean_absolute_error(y_val, predictions)
```

```python
mean_squared_error(y_val,predictions)
```

```python
np.sqrt(mean_squared_log_error(y_val,predictions))
```
